Init

```python
import quick
import flat
import depth
import gen
from visualizer import json_to_stl
from functions import meshify, simplify_mesh, meshfile


stl_path = "mesh/dragon.stl"
out = "out/test"
stl = meshify(stl_path)
stl = simplify_mesh(stl, 1000)
meshfile(stl, out + "_SIMPLIFIED.obj")
fbx = ["models/timber_structures/T_FLOOR.fbx","models/timber_structures/T_TRIFLOOR.fbx","models/timber_structures/T_WALLDIAGONAL.fbx"]
```

quick convert Test

```python
quick.convert(
    target_mesh=stl,
    output_path=out + "_T_FLOOR.json"
)
json_to_stl(out + "_T_FLOOR.json", "models", out +"_T_FLOOR.stl")
```

depth convert Test

```python
depth.convert(
    target_mesh=stl,
    part_fbx_paths=fbx,
    output_path=out + "_depth.json",
)
json_to_stl(out + "_depth.json", "models", out +"_depth.stl")
```

flat convert Test

```python
flat.convert(
    target_mesh=stl,
    output_path=out + "_flat.json",
)
json_to_stl(out + "_flat.json", "models", out +"_flat.stl")
```

gen convert Test

```python
gen.convert(
    target_mesh=stl,
    output_path=out + "_gen.json",
    part_fbx_paths=fbx
)
json_to_stl(out + "_gen.json", "models", out +"_gen.stl")
```
